```python
# ==========================
# Section 1: Upload and Import
# ==========================
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))

# ==========================
# Section 2: Cleaning
# ==========================
df.drop_duplicates(inplace=True)
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Year'] = df['Release_Date'].dt.year

df.fillna({'Country': 'Unknown', 'Type': 'Unknown', 'Category': 'Unknown'}, inplace=True)

# Preprocess genres and countries
genre_df = df[['Type', 'Year']].dropna().copy()
genre_df['Type'] = genre_df['Type'].str.split(', ')
genre_df = genre_df.explode('Type')

country_df = df[['Country']].copy()
country_df['Country'] = country_df['Country'].str.split(', ')
country_df = country_df.explode('Country')

# ==========================
# Section 3: Movies vs TV Shows Over the Years
# ==========================
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="Year", hue="Category", palette="Set2")
plt.title("Movies vs TV Shows Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==========================
# Section 4: Top Genres Trend
# ==========================
top_genres = genre_df['Type'].value_counts().nlargest(8).index
filtered_genres = genre_df[genre_df['Type'].isin(top_genres)]

plt.figure(figsize=(10,5))
sns.countplot(data=filtered_genres, x="Year", hue="Type", palette="muted")
plt.title("Top Genres Trend Over Time")
plt.xticks(rotation=45)
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# ==========================
# Section 5: Top Countries Contributing
# ==========================
top_countries = country_df['Country'].value_counts().nlargest(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="coolwarm")
plt.title("Top 10 Countries Contributing to Netflix")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==========================
# Section 6: Top 10 Directors
# ==========================
top_directors = df['Director'].dropna().value_counts().head(10)

plt.figure(figsize=(8,4))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="magma")
plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.tight_layout()
plt.show()

# ==========================
# Section 7: Movie Duration Distribution
# ==========================
movie_df = df[df['Category'] == 'Movie'].copy()
movie_df['Minutes'] = movie_df['Duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(8,4))
sns.histplot(movie_df['Minutes'], bins=30, kde=True, color='green')
plt.title("Distribution of Movie Durations (in minutes)")
plt.xlabel("Duration")
plt.tight_layout()
plt.show()

# ==========================
# Section 8: TV Show Seasons Count
# ==========================
tv_df = df[df['Category'] == 'TV Show'].copy()
tv_df['Seasons'] = tv_df['Duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(8,4))
sns.countplot(x='Seasons', data=tv_df, palette="Blues", order=tv_df['Seasons'].value_counts().index[:10])
plt.title("Number of Seasons in TV Shows")
plt.tight_layout()
plt.show()
```

```python
# ==========================
# Section 1: Upload and Import
# ==========================
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("📁 Uploading Dataset...")
uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))
print(f"✅ File uploaded! {df.shape[0]} rows, {df.shape[1]} columns loaded.")

# ==========================
# Section 2: Data Cleaning
# ==========================
print("\n🔍 Initial data snapshot:")
print(df.head())

initial_len = len(df)
df.drop_duplicates(inplace=True)
print(f"✅ Dropped {initial_len - len(df)} duplicate rows")

df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Year'] = df['Release_Date'].dt.year

null_counts = df[['Country', 'Type', 'Category']].isna().sum()
print("\n🧼 Nulls before fill:")
print(null_counts)

df.fillna({'Country': 'Unknown', 'Type': 'Unknown', 'Category': 'Unknown'}, inplace=True)
print("✅ Filled missing values with 'Unknown'")

genre_df = df[['Type', 'Year']].dropna().copy()
genre_df['Type'] = genre_df['Type'].str.split(', ')
genre_df = genre_df.explode('Type')
print(f"🎭 Unique genres found: {genre_df['Type'].nunique()}")

country_df = df[['Country']].copy()
country_df['Country'] = country_df['Country'].str.split(', ')
country_df = country_df.explode('Country')
print(f"🌍 Unique countries found: {country_df['Country'].nunique()}")

# ==========================
# Section 3: Movies vs TV Shows Over the Years
# ==========================
print("📊 Plotting: Movies vs TV Shows Over Time")
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="Year", hue="Category", palette="Set2")
plt.title("Movies vs TV Shows Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==========================
# Section 4: Top Genres Trend Over Time
# ==========================
top_genres = genre_df['Type'].value_counts().nlargest(8).index.tolist()
print("🎯 Top 8 Genres:", ', '.join(top_genres))

filtered_genres = genre_df[genre_df['Type'].isin(top_genres)]
plt.figure(figsize=(10,5))
sns.countplot(data=filtered_genres, x="Year", hue="Type", palette="muted")
plt.title("Top Genres Trend Over Time")
plt.xticks(rotation=45)
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# ==========================
# Section 5: Top Countries Contributing
# ==========================
top_countries = country_df['Country'].value_counts().nlargest(10)
print("🌐 Top 10 Countries:")
print(top_countries)

plt.figure(figsize=(8,4))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="coolwarm")
plt.title("Top 10 Countries Contributing to Netflix")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==========================
# Section 6: Top 10 Directors
# ==========================
top_directors = df['Director'].dropna().value_counts().head(10)
print("🎥 Top 10 Directors:")
print(top_directors)

plt.figure(figsize=(8,4))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="magma")
plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.tight_layout()
plt.show()

# ==========================
# Section 7: Movie Duration Distribution (with fix)
# ==========================
movie_df = df[df['Category'] == 'Movie'].copy()
movie_df['Minutes'] = movie_df['Duration'].str.extract(r'(\\d+)').astype(float)
movie_df = movie_df.dropna(subset=['Minutes'])

if not movie_df.empty:
    print("🎞️ Movie duration stats:")
    print(movie_df['Minutes'].describe())

    plt.figure(figsize=(8,4))
    sns.histplot(movie_df['Minutes'], bins=30, kde=True, color='green')
    plt.title("Distribution of Movie Durations (in minutes)")
    plt.xlabel("Duration")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No valid movie durations found to plot.")

# ==========================
# Section 8: TV Show Seasons Count (with fix)
# ==========================
tv_df = df[df['Category'] == 'TV Show'].copy()
tv_df['Seasons'] = tv_df['Duration'].str.extract(r'(\\d+)').astype(float)
tv_df = tv_df.dropna(subset=['Seasons'])

if not tv_df.empty:
    print("📡 Season count frequency:")
    print(tv_df['Seasons'].value_counts().head(10))

    plt.figure(figsize=(8,4))
    sns.countplot(x='Seasons', data=tv_df, palette="Blues", order=tv_df['Seasons'].value_counts().index[:10])
    plt.title("Number of Seasons in TV Shows")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No valid season count data found to plot.")

print("\n✅ All analysis completed successfully!")
```
